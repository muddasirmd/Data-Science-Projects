### Modelling without Pre-processing Text data


```python
#import train-test-split from sklearn 
from sklearn.model_selection import train_test_split
```

```python
#Do the 'train-test' splitting with test size of 20% with random state of 2022 and stratify sampling too
X_train, X_test, y_train, y_test = train_test_split(
    df.Text, 
    df.label_num, 
    test_size=0.2, 
    random_state=2022, 
    stratify=df.label_num)
```

#### Attempt 1 :

1. using sklearn pipeline module create a classification pipeline to classify the Data.


#### Note:

* using CountVectorizer with unigram, bigram, and trigrams.

* use KNN as the classifier with n_neighbors of 10 and metric as 'euclidean' distance.
  
* print the classification report.

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier
```

```python
#1. create a pipeline object
clf_1 = Pipeline([
    ("vectorizer_trigrams", CountVectorizer(ngram_range=(1,3))),
    ("KNN", KNeighborsClassifier(n_neighbors=10, metric='euclidean'))
])
```

```python
#2. fit with X_train and y_train
clf_1.fit(X_train, y_train)
```

```python
#3. get the predictions for X_test and store it in y_pred
y_pred = clf_1.predict(X_test)
```

```python
#4. print the classfication report
print(classification_report(y_test, y_pred))
```

### Use text pre-processing to remove stop words, punctuations and apply lemmatization

```python
import spacy
```

```python
# load english language model and create nlp object from it
nlp = spacy.load("en_core_web_sm") 
```

```python
#use this utility function to get the preprocessed text data
def preprocess(text):
    preprocessText = []
    doc = nlp(text)
    for token in doc:
        if not token.is_stop and not token.is_punct:
            preprocessText.append(token.lemma_)

    return " ".join(preprocessText)
```

```python
df['preprocessed_text'] = df['Text'].apply(preprocess)
```

```python
df.head()
```

#### Build a model with pre processed text


```python
#Do the 'train-test' splitting with test size of 20% with random state of 2020 and stratify sampling too
X_train, X_test, y_train, y_test = train_test_split(
    df.preprocessed_text,
    df.label_num,
    test_size=0.2,
    stratify=df.label_num,
    random_state=2020
)
```
